```python
class User:
    def __init__(self, username, email, password):
        self.username = username
        self.email = email
        self.password = password
    
    def update_email(self, new_email):
        self.email = new_email
    
    def update_password(self, new_password):
        self.password = new_password

# Example Usage
user = User("username", "email@email.com", "1234")
user.update_email("new_email@email.com")
user.update_password("abcd")
print(f"Email: {user.email}, Password: {user.password}")

class Player:
    def __init__(self, name, score=0, health=100):
        self.name = name
        self.score = score
        self.health = health
    
    def increase_score(self, points):
        self.score += points
    
    def decrease_health(self, damage):
        self.health -= damage
    
    def heal(self, amount):
        self.health += amount

# Example Usage
player = Player("Player1")
player.increase_score(10)
player.decrease_health(20)
player.heal(10)
print(f"Score: {player.score}, Health: {player.health}")

class BankAccount:
    def __init__(self, account_number, balance=0):
        self.account_number = account_number
        self.balance = balance
    
    def deposit(self, amount):
        self.balance += amount
    
    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
        else:
            print("Withdrawal failed: insufficient funds.")

# Example Usage
account = BankAccount("12345678", 1000)
account.deposit(500)
account.withdraw(200)
account.withdraw(1500)  # This should fail
print(f"Balance: {account.balance}")
```

```python
class Database:
    connections = 0
    
    @classmethod
    def add_connection(cls):
        cls.connections += 1

# Example Usage
Database.add_connection()
print(f"Connections: {Database.connections}")

class Pizza:
    @classmethod
    def create_margherita(cls):
        return cls("Margherita", ["cheese", "tomato"])
    
    @classmethod
    def create_pepperoni(cls):
        return cls("Pepperoni", ["cheese", "pepperoni"])

    def __init__(self, name, ingredients):
        self.name = name
        self.ingredients = ingredients

# Example Usage
margherita = Pizza.create_margherita()
pepperoni = Pizza.create_pepperoni()
print(f"{margherita.name} - Ingredients: {margherita.ingredients}")
print(f"{pepperoni.name} - Ingredients: {pepperoni.ingredients}")

class Library:
    total_books = 0
    
    @classmethod
    def add_book(cls, count):
        cls.total_books += count
    
    @classmethod
    def get_total_books(cls):
        return cls.total_books

# Example Usage
Library.add_book(5)
print(f"Total Books: {Library.get_total_books()}")
```

```python
class Calculator:
    @staticmethod
    def add(a, b):
        return a + b
    
    @staticmethod
    def subtract(a, b):
        return a - b
    
    @staticmethod
    def multiply(a, b):
        return a * b
    
    @staticmethod
    def divide(a, b):
        if b != 0:
            return a / b
        else:
            return "Cannot divide by zero."

# Example Usage
a, b = 10, 5
print(Calculator.add(a, b))
print(Calculator.subtract(a, b))
print(Calculator.multiply(a, b))
print(Calculator.divide(a, b))

class StringUtils:
    @staticmethod
    def reverse_string(s):
        return s[::-1]
    
    @staticmethod
    def is_palindrome(s):
        return s == s[::-1]

# Example Usage
string = "racecar"
print(StringUtils.reverse_string(string))
print(StringUtils.is_palindrome(string))

class DateUtils:
    @staticmethod
    def is_leap_year(year):
        return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)
    
    @staticmethod
    def days_in_month(year, month):
        if month in [1, 3, 5, 7, 8, 10, 12]:
            return 31
        elif month in [4, 6, 9, 11]:
            return 30
        elif month == 2:
            return 29 if DateUtils.is_leap_year(year) else 28
        else:
            return 0

# Example Usage
year = 2020
month = 2
print(DateUtils.is_leap_year(year))
print(DateUtils.days_in_month(year, month))
```
